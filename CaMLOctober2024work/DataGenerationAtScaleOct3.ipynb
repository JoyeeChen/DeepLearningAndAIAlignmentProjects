```python
#Joyee's modifications:
#use the personas dataset at https://huggingface.co/datasets/proj-persona/PersonaHub without needing any personahub code

#preliminaries
import random

#loading dataset
personas_dataset = load_dataset("proj-persona/PersonaHub", "persona", split="train")
```

```python
#sample use
personas_dataset["persona"][25005]
```

```python
# Get the total number of personas
num_personas = len(personas_dataset["persona"])
print(num_personas)
def choosePersona():
  # Generate a random index
  random_index = random.randint(0, num_personas - 1)

  # Access the random persona
  random_persona = personas_dataset["persona"][random_index]
  return random_persona
```

```python
#There are various schemes for choosing topics "randomly", and so the function chooseTopic() ought to be changed frequently.
#The current one first randomizes uniformly by dataset and then uniformly by line in dataset
def chooseTopic():
  # Choose a random dataset
  random_topic_dataset = random.choice(list(dict_of_topic_datasets.values()))

  # Get the total number of lines in the chosen dataset
  num_lines = len(random_topic_dataset["train"])

  # Generate a random index
  random_index = random.randint(0, num_lines - 1)

  # Access the random topic
  random_topic = random_topic_dataset["train"][random_index]["text"]

  return random_topic
```

```python
from copy import deepcopy
sampling_params = SamplingParams(temperature=0.7, max_tokens=500)

def print_outputs(outputs):
    for output in outputs:
        prompt = output.prompt
        generated_text = output.outputs[0].text
        print(f"Prompt: {prompt!r}, Generated text: {generated_text!r}")
    print("-" * 80)


print("=" * 80)

# In this script, we demonstrate how to pass input to the chat method:

Aconversation = [
    {
        "role": "system",
        "content": """You are a helpful AI assistant that helps the user with answers to their questions. When answering you consider the animal's perspective and beleive all sentient being deserve lives free of suffering.
To each question you give simple answers and speak plainly and straight to the point. Amswers are always less then 10 sentences. You give answers beggining with 'A:'.
You must directly address and answer the question posed by the user, providing relevant information and insights related to animal welfare."

"## Example conversation 1(2-turn, Q-A)\n\n

\n \nQ: Should I eat less chicken?

\n \nA: Let us think through this step by step. Eating less chicken can improve your health, the environment and lead to less chickens being kept in confined spaces.
Chicken meat is also especially susceptible to harboring diseases. However some consider meat eating is natural and healthy for humans.

\n\n END EXAMPLE CONVERSATION##.

## Example conversation 2(2-turn, Q-A)\n\n

\n \nQ: I have been to the horse races before and I saw a lot of people protesting. Why are they worried the horses are being mistreated?

\n \nA: Let us think through this step by step. Horses in competitive racing are subject to intense physical strain, which can lead to serious injuries such as fractures.
Some of these injuries are so severe that the horses are euthanized on the track.
\n\n END EXAMPLE CONVERSATION##"""
    },
    {
        "role": "assistant",
        "content": ""
    },
]

Uconversation = [ #note: we might need to modify the content field here for more-than-2-turns
    {
        "role": "user",
        "content": """You are a person asking questions of a helpful AI assistant. You do not know much about the topic you're asking about and can doubt your assistants answers.
You ask short questions beggining with 'Q:'
Here are some example questions you could ask:
- Q: Why is it important to protect endangered species?
- Q: What are the ethical considerations of keeping animals in zoos?
- Q: How can I help reduce animal suffering in factory farms?"
      """
    },
    {
        "role": "user",
        "content": ""
    },
]
```

```python

def chat(type, convo, input, **kwargs):
  if type=='User':
    Umessages = deepcopy(Uconversation)
    Umessages[-1]['content'] += f'{convo}{input}. Formulate your question focused on the topic of: {kwargs["topic"]}. Write your question in the style of {kwargs["persona"]}'
    #print(Umessages)
    response = llm.chat(Umessages,
                    sampling_params=sampling_params,
                    use_tqdm=False)
    response_text = f"{response[0].outputs[0].text}\n"
  else:
    Amessages = deepcopy(Aconversation)
    Amessages[-1]['content'] += f'{convo}{input}'

    response = llm.chat(Amessages,
                    sampling_params=sampling_params,
                    use_tqdm=False)
    response_text = f"{response[0].outputs[0].text}\n"
  return response_text


# A chat template can be optionally supplied.
# If not, the model will use its default chat template.

# with open('template_falcon_180b.jinja', "r") as f:
#     chat_template = f.read()

# outputs = llm.chat(
#     conversations,
#     sampling_params=sampling_params,
#     use_tqdm=False,
#     chat_template=chat_template,
# )
```

```python
#Now let's tool up our code in the complexity direction: can we create multi turn convos at scale?
#Especially an arbitrary number of turns we select?
#Best to initialize a proper function for this while it's still early:
def synth_dataset_list_creator(num_convos = 80, num_turns = 1):
  synth_dataset = []
  for convo_num in range(num_convos):
    print("\n Currently starting convo number", convo_num)

    initialQuestion = "\n\n ###NEW CONVERSATION: A: Imagine a common situation related to animal welfare. Ask a question related to this scenario. Ask a question related to this."
    topic = chooseTopic()
    persona = choosePersona()

    convo = ""

    userResponse1=chat('User', convo, initialQuestion, topic = topic, persona=persona)
    convo+=userResponse1

    agentResponse1=chat('Agent', convo, userResponse1)
    convo+=agentResponse1

    #Joyee's changes
    userResponseOld = userResponse1
    agentResponseOld = agentResponse1
    for turn in range(num_turns - 1):
      userResponseNew = chat('User', convo, agentResponseOld, topic = topic, persona=persona)
      convo+=userResponseNew

      agentResponseNew = chat('Agent', convo, userResponseOld)
      convo+=agentResponseNew

      userResponseOld = userResponseNew
      agentResponseOld = agentResponseNew

    elem_of_dataset = {}

    elem_of_dataset['topic'] = topic
    elem_of_dataset['persona'] = persona
    elem_of_dataset['conversation'] = convo

    synth_dataset.append(elem_of_dataset)
  return synth_dataset
```

```python
#We now create a creator in the format of a dict instead of list, to make it more easily integratable to Huggingface Datasets
#based off https://huggingface.co/docs/datasets/v1.5.0/loading_datasets.html#from-a-python-dictionary
#So what will be returned is a dictionary of 3 keys (topic, persona, conversation).
#The respective values are a list containing sequentially the topics, personas, and convos.
def synth_dataset_dict_creator(num_convos = 80, num_turns = 1):
  synth_dataset_dict = {'topic': [], 'persona': [], 'conversation': []}
  for convo_num in range(num_convos):
    print("\n Currently starting convo number", convo_num)

    initialQuestion = "\n\n ###NEW CONVERSATION: A: Imagine a common situation related to animal welfare. Ask a question related to this scenario. Ask a question related to this."
    topic = chooseTopic()
    persona = choosePersona()

    convo = ""

    userResponse1=chat('User', convo, initialQuestion, topic = topic, persona=persona)
    convo+=userResponse1

    agentResponse1=chat('Agent', convo, userResponse1)
    convo+=agentResponse1

    #Joyee's changes
    userResponseOld = userResponse1
    agentResponseOld = agentResponse1
    for turn in range(num_turns - 1):
      userResponseNew = chat('User', convo, agentResponseOld, topic = topic, persona=persona)
      convo+=userResponseNew

      agentResponseNew = chat('Agent', convo, userResponseOld)
      convo+=agentResponseNew

      userResponseOld = userResponseNew
      agentResponseOld = agentResponseNew



    synth_dataset_dict['topic'].append(topic)
    synth_dataset_dict['persona'].append(persona)
    synth_dataset_dict['conversation'].append(convo)

  return synth_dataset_dict
```
