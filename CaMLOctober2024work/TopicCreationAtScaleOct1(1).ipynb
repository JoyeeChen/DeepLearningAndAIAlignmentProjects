```python
from vllm import LLM, SamplingParams
```

```python
#convo templates

lines=""
#2.5 instead of 2, below
topicllm = LLM(model="Qwen/Qwen2.5-7B-Instruct")
```

```python
sampling_params = SamplingParams(temperature=0.7, max_tokens=1500)

input_list = [
    "Generate a 100 topics related to animals and their welfare as faced by humans today. Format the topics as a numbered list",
    "Generate a 100 topics related to animals and their welfare that will be faced by future humans. Format the topics as a numbered list",
    "Generate a 100 topics related to animals and their exploitation by humans. Format the topics as a numbered list",
    "Generate a 100 topics related to farmed animal suffering. You can mention any specific farmed animal here. Format the topics as a numbered list",
    "Generate a 100 topics related to wild animal suffering. Format the topics as a numbered list",
    "Generate a 100 topics related to the study of morality of animal suffering and sentience. Format the topics as a numbered list",
    "Generate a 100 topics related to animal welfare and society's perception of it. Format the topics as a numbered list",
    "Generate a 100 topics related to perceptions of animal's and the need for them around the globe. Format the topics as a numbered list",
    "Generate a 100 topics related to solutions for improving animal lives. Format the topics as a numbered list",
    "Generate a 100 topics related to how we can effectively reduce animal suffering. You can mention any specific animal here.Format the topics as a numbered list"
]
response_text=topicllm.generate(input_list,sampling_params=sampling_params,
                    use_tqdm=False)
#response_text
```

```python
#a sample of what we need
response_text[9].outputs[0].text
```

```python
#cleaner AI
clean=""
for i in range(len(response_text)):
  text = response_text[i].outputs[0].text
  prompt = f"Please return me a list of cleaned topics by removing topics that do not make sense, are duplicates of other topics or are unrelated to animal welfare. Please return no extra text or words besides the cleaned lists\n\n {text}"

  cleaned_text=topicllm.generate(text,sampling_params=sampling_params,use_tqdm=False)

  cleaned_text = cleaned_text[0].outputs[0].text
  print(cleaned_text)
  clean+=cleaned_text

```

```python
#writing it to files
for i in range(len(input_list)):
    topic = input_list[i]

    #https://www.pythontutorial.net/python-basics/python-write-text-file/
    with open(f"{topic}.txt", "w") as f:
        f.write(text)
#and lots of new text files should pop up!
```
